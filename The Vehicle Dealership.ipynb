```python
class Vehicle:
    def __init__(self, brand, model, year, price):
        self.brand = brand
        self.model = model
        self.year = year
        self.price = price
        self.is_sold = False
    
    def get_brand(self):
        return self.brand
    
    def get_model(self):
        return self.model
    
    def get_year(self):
        return self.year
    
    def get_price(self):
        return self.price
    
    def check_availability(self):
        return not self.is_sold
    
    def sell(self):
        pass
    
    def __str__(self):
        pass


class Car(Vehicle):
    def __init__(self, brand, model, year, price):
        super().__init__(brand, model, year, price)
    def sell(self):
        if self.is_sold:
            print(f"The {self.brand} {self.model} Car is already sold.")
        else:
            self.is_sold = True
            print(f" The {self.brand} {self.model} Car has been sold.")
    
    def __str__(self):
        return f"Car - Brand: {self.brand}, Model: {self.model}, Year: {self.year}, Price: {self.price}TL\n"


class Motorcycle(Vehicle):
    def __init__(self, brand, model, year, price):
        super().__init__(brand, model, year, price)
        
    def sell(self):
        if self.is_sold:
            print(f"The {self.brand} {self.model} Motorcycle is already sold.")
        else:
            self.is_sold = True
            print(f" The {self.brand} {self.model} Motorcycle has been sold.")
    
    def __str__(self):
        return f"Motorcycle - Brand: {self.brand}, Model: {self.model}, Year: {self.year}, Price: {self.price}TL\n"


class VehicleDealership:
    def __init__(self, name):
        self.name = name
        self.available_vehicles = []
    
    def get_name(self):
        return self.name
    
    def add_vehicle(self, vehicle):
        self.available_vehicles.append(vehicle)
        print(f"Added {vehicle.get_brand()} {vehicle.get_model()} to the dealership.\n")
    
    def remove_vehicle(self, vehicle):
        if vehicle in self.available_vehicles:
            self.available_vehicles.remove(vehicle)
            print(f"Removed {vehicle.get_brand()} {vehicle.get_model()} from the dealership.\n")
        else:
            print(f"The {vehicle.get_brand()} {vehicle.get_model()} is not found in the dealership.\n")
    
    def search_vehicle(self, brand, model):
        found_vehicles = []
        for vehicle in self.available_vehicles:
            if brand.lower() in vehicle.get_brand().lower() and model.lower() in vehicle.get_model().lower():
                found_vehicles.append(vehicle)
        
        if found_vehicles:
            print(f"Found {len(found_vehicles)} vehicle(s) matching the brand '{brand}' and model '{model}':\n")
            for vehicle in found_vehicles:
                if not vehicle.is_sold:
                    print("This Vehicle is available:")
                    print(vehicle)
                if vehicle.is_sold:
                    print("This vehicle is sold.")
                    print(vehicle)
        else:
            print(f"No vehicles found matching the brand '{brand}' and model '{model}'.\n")
    
    def display_available_vehicles(self):
        print("The vehicles in the dealership:\n")
        for vehicle in self.available_vehicles:
            if not vehicle.is_sold:
                print("This Vehicle is available:")
                print(vehicle)
            if vehicle.is_sold:
                print("This vehicle is sold.")
                print(vehicle)
    
    def sell_vehicle(self, brand, model):
        vehicle = None
        for available_vehicle in self.available_vehicles:
            if available_vehicle.get_brand().lower() == brand.lower() and available_vehicle.get_model().lower() == model.lower():
                vehicle = available_vehicle
                break
        
        if vehicle:
            vehicle.sell()
        else:
            print("Vehicle not found.\n")


def display_menu():
    print("Welcome to the Vehicle Dealership\n")
    print("1. Add a vehicle to the dealership")
    print("2. Remove a vehicle from the dealership")
    print("3. Display available vehicles in the dealership")
    print("4. Search for a vehicle")
    print("5. Sell a vehicle")
    print("else :  Exit")



dealership = VehicleDealership("The Vehicle Dealership")

while True:
    display_menu()
    choice = input("What is Your Option")

    if choice == "1":
        brand = input("Enter the brand of the vehicle: ")
        model = input("Enter the model of the vehicle: ")
        year = int(input("Enter the year of the vehicle: "))
        price = float(input("Enter the price of the vehicle: "))
        vehicle_type = input("Enter the type (car/motorcycle): ")

        if vehicle_type == "car":
            vehicle = Car(brand, model, year, price)
        elif vehicle_type == "motorcycle":
            vehicle = Motorcycle(brand, model, year, price)
        else:
            print("Invalid vehicle type.")
            continue

        dealership.add_vehicle(vehicle)

    elif choice == "2":
        brand = input("Enter the brand  ")
        model = input("Enter the model ")
        vehicle = None
        for available_vehicle in dealership.available_vehicles:
            if available_vehicle.get_brand().lower() == brand.lower() and available_vehicle.get_model().lower() == model.lower():
                vehicle = available_vehicle
                break
        
        if vehicle:
            dealership.remove_vehicle(vehicle)
        else:
            print("Vehicle not found.")

    elif choice == "3":
        dealership.display_available_vehicles()

    elif choice == "4":
        brand = input("Enter the brand ")
        model = input("Enter the model ")
        dealership.search_vehicle(brand, model)

    elif choice == "5":
        brand = input("Enter the brand ")
        model = input("Enter the model ")
        dealership.sell_vehicle(brand, model)

    else:
        break
        

print("Thank you for using the Vehicle dealership program!")
```
